# DATA CLEANING

```python
# Filling Missing Values In Age

df['Age']=df['Age'].fillna(df['Age'].median())
df.isna().sum()
```

```python
# Filling Missing Values In Embarked

df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode()[0])
df.isnull().sum()
```

```python
# Filling Missing Values In Cabin

df['Cabin']=df['Cabin'].fillna('NoCabin')
df.isnull().sum()
```

# DATA VISULALIZATION (EDA)


```python
# Importing Data Visualization Modules

import matplotlib.pyplot as plt
import seaborn as sns
```

Different cmap colors:


*   rocket
*   mako
*   flare
*   crest
*   magma
*   viridis
*   rocket_R



```python
# Correlation Heatmap Of Numeric Columns

plt.figure(figsize=(7,5))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='mako')
plt.show()
```

```python
# Correlation Heatmap Of Survival

sns.heatmap(df.corr(numeric_only=True)[['Survived']],annot=True,cmap='rocket_r')
plt.show()
```

```python
# Survival Count

plt.figure(figsize=(2,2))
sns.countplot(x='Survived', data=df, palette=['red','lightgreen'],edgecolor='black')
plt.ylabel('Number Of Passengers')
plt.xticks([0,1],['Did Not Survive','Survived'])
plt.title('Survival Count')
plt.show()
```

```python
# Survival By Gender

plt.figure(figsize=(3,3))
colors={'male':'lightblue','female':'pink'}
sns.countplot(x='Survived',hue='Sex',data=df,palette=colors)
plt.title("Survival by Gender")
plt.ylabel('Number Of Passengers')
plt.xticks([0,1],['Did Not Survive','Survived'])
plt.show()
```

```python
# Survival By Passenger Class

plt.figure(figsize=(3,3))
sns.countplot(x='Survived',hue='Pclass',data=df)
plt.title('Survival By Passenger Class')
plt.ylabel('Number Of Passengers')
plt.xticks([0,1],['Did Not Survive','Survived'])
plt.show()
```

```python
# Age Distribution By Survival

plt.figure(figsize=(3,3))
sns.displot(x='Age',data=df,hue='Survived',kind='kde',fill=True,height=3,aspect=1.5,palette={0:'pink',1:'green'})
plt.show()
```

```python
# Survival by Embarked Port

plt.figure(figsize=(3,3))
sns.countplot(x='Embarked',hue='Survived',data=df,palette={0:'red',1:'lightgreen'})
plt.show()
```

```python
# Boxplot Of Fare vs PClass

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

palette = ['pink', 'lightgreen', 'lightblue']

sns.boxplot(x='Pclass', y='Fare', data=df, palette=palette, ax=axes[1])
axes[0].set_ylim(0, 150)  # Limit for visibility
axes[0].set_title("Fare by Class (Linear Scale, Zoomed In)")
axes[0].set_xlabel("Passenger Class")
axes[0].set_ylabel("Fare")

sns.boxplot(x='Pclass', y='Fare', data=df, palette=palette, ax=axes[0])
axes[1].set_yscale('log')
axes[1].set_title("Fare by Class (Log Scale)")
axes[1].set_xlabel("Passenger Class")
axes[1].set_ylabel("Fare")

plt.show()
```

```python
# PAIR PLOT (For Multiple Visual Patterns)

sns.pairplot(df[['Survived','Sex','Age','Pclass','Fare']],hue='Survived',palette={0:'red',1:'green'})
plt.show()
```
