# **Spam Email Detector**

**Import Dependencies**

```python
#Import Dependencies
import pandas as pd
from sklearn.metrics import accuracy_score,f1_score,recall_score,confusion_matrix,precision_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

```

**Load the dataset**

```python
df=pd.read_csv("spambase.csv")
```

**First Five values of the dataset**

```python
df.head()
```

**No. of rows and column of the dataset**

```python
df.shape
```

**Statistical measures**

```python
df.describe()
```

**Vertical and horizantal split of the dataset for testing the model**

```python
X=df.drop(columns='spam',axis=1)
y=df['spam']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

**Train the model**

```python
pipeline = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
pipeline.fit(X_train, y_train)
```

**Model prediction based on train data out of 1**

```python
X_train_predict=pipeline.predict(X_train)
X_train_accuracy=accuracy_score(X_train_predict,y_train)
print("Train_Accuracy:",X_train_accuracy)
```

**Model prediction on test data out of 1**

```python
X_test_predict=pipeline.predict(X_test)
X_test_accuracy=accuracy_score(X_test_predict,y_test)
print("Test Accuracy:",X_test_accuracy)
```

## **Classification Report**

```python
print("Classification Report.....")
#Finding Accuracy
X_test_predict=pipeline.predict(X_test)
accuracy=accuracy_score(y_test,X_test_predict)
print("Accuracy Score:",accuracy)
#Finding Precision Score
precision=precision_score(y_test,X_test_predict)
print("Precision Score:",precision)
#Finding Recall Score
recall=recall_score(y_test,X_test_predict)
print("Recall Score:",recall)
#Finding f1 score
f1=f1_score(y_test,X_test_predict)
print("F1 Score:",f1)
#Confusion Matrix
cm=confusion_matrix(y_test,X_test_predict)
sns.heatmap(cm,annot=True)
plt.title("Confusion Matrix")
plt.show()
```

**Classification Report explanation **

Means that out of all the actual positives (class 1), 89.80% were correctly identified.

The harmonic mean of precision and recall. It balances the two, giving a better measure when you have imbalanced classe
