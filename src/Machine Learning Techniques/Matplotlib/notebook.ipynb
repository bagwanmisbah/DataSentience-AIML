# we need to draw both chart side by side


# so y data on chart should have position at x axis  so firstly we take positions of that
using np.arrange((len(x_data))   it gives [0,1,2,3.....] array which is the position of y data onn x axis

now we need to chart side by side so we need width and first bar will be at position+-width and second also similar
this shifts the bar left right



```python
import numpy as np

ages=[25,26,27,28,29,30]   # on x axis

python_dev=[450,567,890,190,300,1000]  #on y axis

java_dev=[900,190,467,890,456,550]  # on y axis


position=np.arange(len(ages))

width=0.3
# Create a figure with a specific size (width, height in inches)
plt.figure(figsize=(10, 6))

plt.bar(position,python_dev,color='b',width=0.35,label="Python developer")

plt.bar(position+width,java_dev,color='g',width=0.35,label="Java developer")  # it will be at right side of first bar
plt.xlabel("Age")
plt.ylabel("Developer's Number")
plt.title("Age vs Developer")
plt.legend()
plt.tight_layout()
plt.show()
```

## PIE CHART

# A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions.

```python

slices=[8999,1235,9000,5789]

labels=['c++','Python','JAva','JS']

plt.pie(slices,labels=labels)
plt.title("PROGRAMMING LANGUAGE")
plt.show()
```

```python
# cahnge the slice color

slices=[8999,1235,9000,5789]

labels=['c++','Python','JAva','JS']

color=['orange','blue','black','green']
plt.pie(slices,labels=labels,colors=color)
plt.title("PROGRAMMING LANGUAGE")
plt.show()
```

```python
# Apply wedge properties
slices=[8999,1235,9000,5789]

labels=['c++','Python','JAva','JS']

color=['orange','blue','black','green']
plt.pie(slices,labels=labels,colors=color,wedgeprops={'edgecolor':'red'})
plt.title("PROGRAMMING LANGUAGE")
plt.show()
```

```python
# Explode means pull out a slice from the pie chart

slices=[8999,1235,9000,5789]

labels=['c++','Python','JAva','JS']

explode=[0,0,0.3,0]    # this pull out java froom the pie chart

color=['orange','blue','gray','green']
plt.pie(slices,labels=labels,colors=color,explode=explode,wedgeprops={'edgecolor':'red'})
plt.title("PROGRAMMING LANGUAGE")
plt.show()
```

```python
# shadow argument show the shadow of piechart
# we want to rotate the pie chart then pass startangle =value
# we want to  show the which slice takke how many % area then pass autopct='%1.1f%%'

slices=[8999,1235,9000,5789]

labels=['c++','Python','JAva','JS']

explode=[0,0,0.3,0]    # this pull out java froom the pie chart

color=['orange','blue','gray','green']
plt.pie(slices,labels=labels,colors=color,explode=explode,wedgeprops={'edgecolor':'red'},shadow=True,startangle=90,autopct='%1.1f%%')
plt.title("PROGRAMMING LANGUAGE")
plt.show()
```

```python
ages=[23,24,25,26,27,28]

py_dev=[34,56,90,100,90,178]

plt.plot(ages,py_dev,color='red',linestyle='--',label='Python')

plt.fill_between(ages,py_dev,color='yellow',alpha=0.3,interpolate=True)
plt.title("AGE VS DEVELOPER")
plt.xlabel("AGE")
plt.ylabel("DEVELOPER")
plt.legend()
plt.show()
```

```python
# FOR 2 DATA

ages=[23,24,25,26,27,28]

py_dev=[34,56,90,100,90,178]

java_dev=[44,16,100,120,80,176]
plt.plot(ages,py_dev,color='red',linestyle='--',label='Python')
plt.plot(ages,java_dev,color='blue',linestyle='-',label='java')

plt.fill_between(ages,py_dev,color='yellow',alpha=0.3,interpolate=True)
plt.fill_between(ages,java_dev,color='green',alpha=0.3,interpolate=True)

plt.title("AGE VS DEVELOPER")
plt.xlabel("AGE")
plt.ylabel("DEVELOPER")
plt.legend()
plt.show()
```

```python
# FOR 2 DATA
import numpy as np
ages=[23,24,25,26,27,28]

py_dev=[34,56,90,100,90,178]

java_dev=[44,16,100,120,80,176]

py_dev = np.array(py_dev)
java_dev = np.array(java_dev)


plt.plot(ages,py_dev,color='red',linestyle='--',label='Python')
plt.plot(ages,java_dev,color='blue',linestyle='-',label='java')

plt.fill_between(ages,py_dev,color='yellow',alpha=0.3,interpolate=True,
                 where=(py_dev<=java_dev))     
# For compare two elements  where condition  you need to convert list into array

plt.title("AGE VS DEVELOPER")
plt.xlabel("AGE")
plt.ylabel("DEVELOPER")
plt.legend()
plt.show()
```

# Now lets see How we can plot figure for ML work


```python
# Losss curve : loss curve is a plot that shows how your model’s loss value changes over training epochs
epochs = range(1, 11)
train_loss = [0.9, 0.8, 0.6, 0.5, 0.45, 0.4, 0.38, 0.35, 0.33, 0.3]
val_loss = [1.0, 0.85, 0.7, 0.6, 0.55, 0.53, 0.52, 0.51, 0.5, 0.49]

plt.figure(figsize=(5,5))
plt.plot(epochs, train_loss, label="Training Loss")
plt.plot(epochs, val_loss, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curves")
plt.legend()
plt.show()
```

```python
# A confusion matrix shows the counts of correct and incorrect predictions for each class,
# helping us see where the model is performing well and where it is making mistakes.”

# Use for classification problem

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = [0, 0, 1, 1, 0, 1, 0, 1, 1, 0]
y_pred = [0, 1, 1, 1, 0, 1, 0, 0, 1, 1]

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```
